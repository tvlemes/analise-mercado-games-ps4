## Mercado de Games

Dados de vendas e classificações de jogos de videogame extraídos do VzCharts

```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv('PS4_GamesSales.csv', encoding='latin-1')

# Verificando
Base_Dados.head()
```

```python
# Verificando as Dimensões 
Base_Dados.shape
```

```python
# Verificando os campos nulos
Base_Dados.isnull().sum()
```

```python
# Gráfico dos campos nulos
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,7))
plt.title('Campos Nulos', fontsize=15, loc='center')

sns.heatmap(Base_Dados.isnull(), cbar=False);
```

```python
# Retirando os valores nulos
Base_Dados.dropna( inplace=True)
```

```python
# Verificando Estatísticas
Base_Dados.describe()
```

```python
# Gráfico de valor vendido por ano
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,6))
plt.title('Quantidade de vendas Globais (mi)', fontsize=15, loc='left')

sns.barplot(data=Base_Dados, x='Year', y='Global', ci=None, color='#69b3a2', estimator=sum);
plt.xlabel('Ano', fontsize=15)
plt.ylabel('Quantidade de vendas (mi)', fontsize=15)
```

```python
# Retirando os anos vazios

Base_Dados = Base_Dados.loc[ ( Base_Dados['Year'] != 2019 ) & ( Base_Dados['Year'] != 2020 ) ]
Base_Dados
```

```python
# Gráfico de Densidade

# Tamanho
plt.figure(figsize=(15,5))

# Titulo
plt.title('Distribuição das vendas Globais', fontsize=15, loc='left')

# Estilo
plt.style.use('ggplot')

# Plot
sns.kdeplot( Base_Dados['Global'], shade=True, bw=1, color='#96a8a8', linewidth=2.5 );
```

```python
# Análise por ano
Base_Dados.groupby( by=['Year'] ).sum()
```

```python
# Gráfico Boxplot

plt.figure( figsize=(16,5) )
plt.title('Análise da distribuição Global (mi)')

sns.boxplot( data=Base_Dados, x='Year', y='Global' );
```

```python
# 10 jogos que mais venderam
Base_Dados.loc[ Base_Dados['Global'] >= 10 ]
```

```python
Analise = Base_Dados.groupby( by=['Year'] ).sum().reset_index()

America = [ America / Total * 100 for America, Total in zip( Analise['North America'], Analise['Global'] ) ]
Europa = [ Europa / Total * 100 for Europa, Total in zip( Analise['Europe'], Analise['Global'] ) ]
Japao = [ Japao / Total * 100 for Japao, Total in zip( Analise['Japan'], Analise['Global'] ) ]
Mundo = [ Mundo / Total * 100 for Mundo, Total in zip( Analise['Rest of World'], Analise['Global'] ) ]

America, Europa, Japao, Mundo
```

```python
# Gráfico empilhado

# Tamanho 
plt.figure(figsize=(14,5))

# Largura da barra
Largura_Barra=0.55
Rotulos = Analise['Year']
Grupos = [0 ,1, 2, 3, 4, 5]

# Titulo
plt.title('Análise distribuição por continentes')

# Plot da America
plt.bar( Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white' );

# Plot da Europa
plt.bar( Grupos, Europa, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white' );

# Plot da Japao
plt.bar( Grupos, Japao, bottom=[ A + B for A, B in zip(America, Europa) ], width=Largura_Barra, color='#a3acff', edgecolor='white' );

# Plot da Mundo
plt.bar( Grupos, Mundo, bottom=[ A + B + C for A, B, C in zip(America, Europa, Japao) ], width=Largura_Barra, color='#d3acfe', edgecolor='white' );

# Labels
plt.xticks( Grupos, Rotulos)
plt.xlabel( 'Grupos' )
plt.ylabel( 'Distribuição %' )

# Legendas
plt.legend(['America N', 'Europa', 'Japão', 'Mundo'], loc='upper center', bbox_to_anchor=(0.15, -0.1), ncol=4);
```

```python
# Transformando String em Numerico
from sklearn.preprocessing import LabelEncoder

Funcao_Label = LabelEncoder()

Base_Dados['Produtor'] = Funcao_Label.fit_transform( Base_Dados['Publisher'] )
Base_Dados['Genero'] = Funcao_Label.fit_transform( Base_Dados['Genre'] )
Base_Dados['Jogo'] = Funcao_Label.fit_transform( Base_Dados['Game'] )

Base_Dados.head()
```

```python
# Paleta de cores
Paleta_Cores = sns.color_palette('husl', 8)
Paleta_Cores
```

```python
# Gráfico Análise dos produtores de Games
plt.figure(figsize=(22,5))
plt.title('Análise dos produtores de Games (mi)')
sns.scatterplot(data=Base_Dados, x='Produtor', y='Global', color=Paleta_Cores[4]);
```

```python
# Gráfico Análise dos generos de Games
plt.figure(figsize=(22,5))
plt.title('Análise dos generos de Games (mi)')
sns.scatterplot(data=Base_Dados, x='Genero', y='Global', color=Paleta_Cores[6]);
```

```python
# Gráfico Análise dos Games
plt.figure(figsize=(22,5))
plt.title('Análise dos Games (mi)')
sns.scatterplot(data=Base_Dados, x='Jogo', y='Global', color=Paleta_Cores[6]);
```
